```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('half_dataset.zip', header=[0, 1])
```

```python
data
```

### График с секндами по оси x

```python
fig = plt.figure(figsize = (25, 15))
ax = fig.add_subplot(111)
ax.fill_between(data['timestamp_second']['Unnamed: 0_level_1'],
                 data['bid_price_0']['mean'], 
                 label = "bid_price_0(mean)", 
                 linewidth = 1
                )
ax.plot(data['timestamp_second']['Unnamed: 0_level_1'],
                data['ask_price_0']['mean'], 
                label = "ask_price_0(mean)", 
                color = 'r')
ax.set_xlabel('Time, s', fontsize = 20)
ax.set_ylabel('Price, $', fontsize = 20)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.set_ylim(min(min(data['bid_price_0']['mean']), min (data['ask_price_0']['mean'])))
ax.set_xlim(min(data['timestamp_second']['Unnamed: 0_level_1']))
ax.legend(prop={'size': 20})
```

### График с годами, но мне кажется, это некорректно
